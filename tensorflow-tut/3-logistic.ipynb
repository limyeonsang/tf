keras의 출력층에서는 z=XW+b로 선형회귀를 계산하고 sigmoid(z)를 연산을 수행함으로써 logistic regression을 keras로 구현한다. 

```python
import numpy as np

try:
    data = np.loadtxt('./resource/pima-indians-diabetes.csv', delimiter=',')

    x_data = data[:, 0:-1]
    y_data = data[:, -1]
    print(x_data.shape, y_data.shape)

except Exception as err:
    print(str(err))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

model = Sequential()
model.add(Dense(1, activation='sigmoid', input_shape=(x_data.shape[1], )))
# 1 = y_data.shape[1]
```

```python
model.compile(optimizer=SGD(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
hist = model.fit(x_data, y_data, epochs=200, validation_split=0.2, verbose=2)
```

```python
model.evaluate(x_data, y_data)
```
